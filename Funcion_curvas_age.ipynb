```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Function

```python
def plot_curve_age_abs(iso2,column):

  path = '/content/drive/MyDrive/Colab Notebooks/'     # change this path according to the database
  country = iso2
  end ='-estimate.csv'

  df = pd.read_csv(path+country+end)

  df_no77 = df[df['age'] != -77]
  df_ok = df_no77[df_no77['age'] != -99]

  y1 = df_ok[df_ok['age'] == 1]
  y2 = df_ok[df_ok['age'] == 2]
  y3 = df_ok[df_ok['age'] == 3]
  y4 = df_ok[df_ok['age'] == 4]
  y5 = df_ok[df_ok['age'] == 5]
  y6 = df_ok[df_ok['age'] == 6]
  y7 = df_ok[df_ok['age'] == 7]

  rng= y1['date'].to_numpy()

  df_print = pd.DataFrame({'18-24':y1[column].to_numpy(),
                   '25-34':y2[column].to_numpy(),
                   '35-44':y3[column].to_numpy(),
                   '45-54':y4[column].to_numpy(),
                   '55-64':y5[column].to_numpy(),
                   '65-74':y6[column].to_numpy(),
                   '>75':y7[column].to_numpy()}, index=rng)
  
  plt.figure(figsize=(10,10))
  plt.plot(df_print.index, df_print['>75'], label = ">75",linewidth=3)
  plt.plot(df_print.index, df_print['65-74'], label = "65-74",linewidth=3)
  plt.plot(df_print.index, df_print['55-64'], label = "55-64",linewidth=3)
  plt.plot(df_print.index, df_print['45-54'], label = "45-54",linewidth=3)
  plt.plot(df_print.index, df_print['35-44'], label = "35-44",linewidth=3)
  plt.plot(df_print.index, df_print['25-34'], label = "25-34",linewidth=3)
  plt.plot(df_print.index, df_print['18-24'], label = "18-24",linewidth=3)
  plt.title(iso2)
  plt.xlabel('Date')
  plt.ylabel(column)
  plt.legend(loc="upper left")
  return
```

# Function

```python
def plot_curve_age_norm(iso2,column):

  path = '/content/drive/MyDrive/Colab Notebooks/'   # change this path according to the database
  country = iso2
  end ='-estimate.csv'

  df = pd.read_csv(path+country+end)

  df_no77 = df[df['age'] != -77]
  df_ok = df_no77[df_no77['age'] != -99]

  y1 = df_ok[df_ok['age'] == 1]
  y2 = df_ok[df_ok['age'] == 2]
  y3 = df_ok[df_ok['age'] == 3]
  y4 = df_ok[df_ok['age'] == 4]
  y5 = df_ok[df_ok['age'] == 5]
  y6 = df_ok[df_ok['age'] == 6]
  y7 = df_ok[df_ok['age'] == 7]

  rng= y1['date'].to_numpy()

  df_print = pd.DataFrame({'18-24':y1[column].to_numpy(),
                   '25-34':y2[column].to_numpy(),
                   '35-44':y3[column].to_numpy(),
                   '45-54':y4[column].to_numpy(),
                   '55-64':y5[column].to_numpy(),
                   '65-74':y6[column].to_numpy(),
                   '>75':y7[column].to_numpy()}, index=rng)
  normalizedDataFrame = df_print.div(df_print.sum(axis=1), axis=0)
  rng= y1['date'].to_numpy()
  normalizedDataFrame.index = rng

  plt.figure(figsize=(10,10))
  plt.plot(normalizedDataFrame.index, normalizedDataFrame['>75'], label = ">75",linewidth=3)
  plt.plot(normalizedDataFrame.index, normalizedDataFrame['65-74'], label = "65-74",linewidth=3)
  plt.plot(normalizedDataFrame.index, normalizedDataFrame['55-64'], label = "55-64",linewidth=3)
  plt.plot(normalizedDataFrame.index, normalizedDataFrame['45-54'], label = "45-54",linewidth=3)
  plt.plot(normalizedDataFrame.index, normalizedDataFrame['35-44'], label = "35-44",linewidth=3)
  plt.plot(normalizedDataFrame.index, normalizedDataFrame['25-34'], label = "25-34",linewidth=3)
  plt.plot(normalizedDataFrame.index, normalizedDataFrame['18-24'], label = "18-24",linewidth=3)
  plt.title(iso2)
  plt.xlabel('Date')
  plt.ylabel(column)
  plt.legend(loc="upper left")
  return 
```

```python
plot_curve_age_abs('ES','p_cliWHO_14days')
```

```python
plot_curve_age_norm('ES','p_cliWHO_14days')
```
