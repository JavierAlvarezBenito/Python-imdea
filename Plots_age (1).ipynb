```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# SPAIN

```python
df_ES = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ES-estimate.csv')
```

```python
df_ES.head()
```

```python
df_no77 = df_ES[df_ES['age'] != -77]

df_ok = df_no77[df_no77['age'] != -99]
```

```python
df_ok.info()
```

```python
df_ok.age.unique()
```

```python
y1 = df_ok[df_ok['age'] == 1]
y2 = df_ok[df_ok['age'] == 2]
y3 = df_ok[df_ok['age'] == 3]
y4 = df_ok[df_ok['age'] == 4]
y5 = df_ok[df_ok['age'] == 5]
y6 = df_ok[df_ok['age'] == 6]
y7 = df_ok[df_ok['age'] == 7]

y1 = y1[["date","age","p_cli_14days","p_cliWHO_14days","p_cli_local_14days"]]
y2 = y2[["date","age","p_cli_14days","p_cliWHO_14days","p_cli_local_14days"]]
y3 = y3[["date","age","p_cli_14days","p_cliWHO_14days","p_cli_local_14days"]]
y4 = y4[["date","age","p_cli_14days","p_cliWHO_14days","p_cli_local_14days"]]
y5 = y5[["date","age","p_cli_14days","p_cliWHO_14days","p_cli_local_14days"]]
y6 = y6[["date","age","p_cli_14days","p_cliWHO_14days","p_cli_local_14days"]]
y7 = y7[["date","age","p_cli_14days","p_cliWHO_14days","p_cli_local_14days"]]
```

### p_cli_14days Spain

```python
rng= y1['date'].to_numpy()

df_print = pd.DataFrame({'18-24':y1['p_cli_14days'].to_numpy(),
                   '25-34':y2['p_cli_14days'].to_numpy(),
                   '35-44':y3['p_cli_14days'].to_numpy(),
                   '45-54':y4['p_cli_14days'].to_numpy(),
                   '55-64':y5['p_cli_14days'].to_numpy(),
                   '65-74':y6['p_cli_14days'].to_numpy(),
                   '>75':y7['p_cli_14days'].to_numpy()}, index=rng)
```

```python
from matplotlib.font_manager import FontProperties
fontP = FontProperties()
fontP.set_size('xx-small')
```

```python
plt.figure(figsize=(10,10))
plt.stackplot(df_print.index, df_print['18-24'], df_print['25-34'], df_print['35-44'],df_print['45-54'],df_print['55-64'],df_print['65-74'],df_print['>75'], labels=df_print.columns)
plt.title('SPAIN')
plt.xlabel('Date')
plt.ylabel('p_cli_14days')
plt.legend(loc="upper left")
plt.show()
```

```python
normalizedDataFrame = df_print.div(df_print.sum(axis=1), axis=0)


rng= y1['date'].to_numpy()
normalizedDataFrame.index = rng
```

```python
plt.figure(figsize=(10,10))
plt.stackplot(normalizedDataFrame.index, normalizedDataFrame['18-24'], normalizedDataFrame['25-34'], normalizedDataFrame['35-44'],normalizedDataFrame['45-54'],normalizedDataFrame['55-64'],normalizedDataFrame['65-74'],normalizedDataFrame['>75'], labels=normalizedDataFrame.columns)
plt.title('SPAIN')
plt.xlabel('Date')
plt.ylabel('p_cli_14days')
plt.legend(loc="upper left")
plt.show()
```
